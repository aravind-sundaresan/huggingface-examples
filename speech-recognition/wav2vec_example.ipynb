```python
import librosa
import torch
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
```

```python
# Reading sample audio clip
import IPython.display as display

display.Audio("data/shawshank_redemption_clip.wav")
```

```python
# Loading the audio file into a float array 
audio, sr = librosa.load("data/shawshank_redemption_clip.wav", sr=16000)
```

```python
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
# Converting the audio array into a PyTorch tensor
input_values = tokenizer(audio, return_tensors="pt").input_values
```

```python
input_values
```

```python
logits = model(input_values).logits
predictions = torch.argmax(logits, dim=-1)
```

```python
# Decoding model output using a tokenizer
transcription = tokenizer.batch_decode(predictions)[0]
transcription
```
